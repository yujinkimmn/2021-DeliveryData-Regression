# Mulitiple Linear Regression 실행

```python
#교안에 있는 multiple linear regression 

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 


scaler = StandardScaler()
```

```python
X = df.iloc[:, :4]
X.head()
```

```python
Y = df[['time']]
Y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=1)
```

```python
scaler.fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
mlr = LinearRegression()
```

```python
mlr.fit(X_train_scaled, y_train)
```

```python
y_pred = mlr.predict(X_test_scaled)
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel("actual time")
plt.ylabel("predicted time")
plt.xlim(0,60)
plt.ylim(0,60)
```

```python
# residual plots
def residual_plot(y_pred, y):
    plt.scatter(y_pred, y_pred-y, c="blue", marker='s', label="Residual plot")
    plt.xlabel("Predicted values")
    plt.ylabel("Residuals")
    plt.hlines(y=0, color="black", xmin=-10, xmax=50)
```

```python
residual_plot(y_pred, y_test)
```

```python
mlr.score(X_test, y_test)

#오류가 너무 크다 선형 회귀가 아니라 다항회귀로 해보자(Polynomial Regression)
```
